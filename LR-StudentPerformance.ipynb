```python
# Insert the data

import pandas as pd
df = pd.read_csv("students_performance.csv")
df
```

```python
# Split the X and the y dataset
from sklearn.model_selection import train_test_split

X = df["exam_score"]
y = df.iloc[:, [1, 4, 5, 6, 8, 9, 11, 14, 16, 18]]

# Train and Test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
y_train = pd.DataFrame(y_train)
y_test = pd.DataFrame(y_test)
```

```python
# Model instantiated from the imported Model class
from sklearn import linear_model

lr_model = linear_model.LinearRegression()
lr_model = lr_model.fit(X_train, y_train)
```

```python
# Predict the model

y_pred = lr_model.predict(X_test)
```

```python
# Compare ethe y_pred and the actual y_train
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np

model_rmse = np.sqrt(mean_absolute_error(y_test, y_pred))
model_r2 = r2_score(y_test, y_pred)

print(model_rmse)
print(model_r2)
```
